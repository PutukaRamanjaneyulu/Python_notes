**Matplotlib :**

- Scatter plot
- Line Plot or Line Chart
- Legend Function
- Subplots
- Histogram
- Bar_Charts
- Pie Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Scatter Plot : Support's Only 2D Representation of the data
a=[4,1,8,7,9,3,2]
b=[5,3,1,4,6,2,9]
plt.scatter(x=a,y=b,color='r')
```

```python
plt.figure(figsize=(5,4))
plt.title("Scatter plot")
plt.xlabel('Values_of_a')
plt.ylabel('Values_of_b')
a=[4,1,8,7,9,3,2]
b=[5,3,1,4,6,2,9]
plt.scatter(x=a,y=b,color='r',marker='s')
```

```python
# line Chart : Line cahrt is nothing But Connecting Scatter points is called Line Chart

plt.figure(figsize=(5,4))
plt.title("Line cahrt")
plt.xlabel('Values_a')
plt.ylabel('Values_a')
a=[4,1,8,7,9,3,2]
b=[5,3,1,4,6,2,9]
plt.plot(a,b,color='g')
```

```python
# Issue With the line chart: Multiple lines come in one diagram | we can slove these issues using the sub plots
plt.figure(figsize=(5,4))
india=[4,1,8,7,9,3,2]
usa=[5,3,1,4,6,2,9]
uk=[3,2,5,7,1,4,9]
plt.plot(india, usa,color = 'r' , marker ='*')
plt.plot(india,uk,color='g', marker='d')
plt.show()
```

```python
plt.figure(figsize=(5,4))
india=[4,1,8,7,9,3,2]
usa=[5,3,1,4,6,2,9]
uk=[3,2,5,7,1,4,9]
plt.plot(india, usa,color = 'r' , marker ='*',label='India_usa')
plt.plot(india,uk,color='g', marker='d', label='India_Uk')
plt.legend()
plt.show()
```

```python
# Subplots # Row wise Subplots
plt.figure(figsize=(5,4))
india=[4,1,8,7,9,3,2]
usa=[5,3,1,4,6,2,9]
uk=[3,2,5,7,1,4,9]
plt.subplot(1,2,1) #Row , Box , Number of boxes
plt.plot(india,usa,color='r',marker='*',label="India_usa")
plt.legend()
plt.subplot(1,2,2)
plt.plot(india,uk,color="g",marker='d',label='India_uk')
plt.legend()
plt.show()
```

```python
# Subplots # Coloumn wise Subplots
plt.figure(figsize=(5,4))
india=[4,1,8,7,9,3,2]
usa=[5,3,1,4,6,2,9]
uk=[3,2,5,7,1,4,9]
plt.subplot(2,1,1) #Row , Box , Number of boxes
plt.plot(india,usa,color='r',marker='*',label="India_usa")
plt.legend()
plt.subplot(2,1,2)
plt.plot(india,uk,color="g",marker='d',label='India_uk')
plt.legend()
plt.show()
```

```python
# Histogram : Histogram is 1D in between the intervels what will happen
plt.figure(figsize=(6,4))
age=[20,10,15,80,70,30]
plt.hist(age,bins=4)
plt.show()
```

```python
# Barcharts : Virtical way representation
plt.figure(figsize=(6,7))
plt.title("Bar_Chart_V")
plt.xlabel('Company')
plt.ylabel('salary')
Company=['Google','meta','msft','IBM']
salary=[1600,1200,1400,800]
plt.bar(Company,salary,color=['r','g','b','y'])
plt.show()
```

```python
# Barcharts : Horizontal way representation
plt.figure(figsize=(6,7))
plt.title("Bar_Chart_V")
plt.xlabel('Company')
plt.ylabel('salary')
Company=['Google','meta','msft','IBM']
salary=[1600,1200,1400,800]
plt.barh(Company,salary,color=['r','g','b','y'])
plt.show()
```

```python
# Pie Chart : Pie chart means Circle Based on the values
plt.figure(figsize=(6,8))
plt.title("Pie_Chart")
Company=['Google','meta','msft','IBM']
salary=[1600,1200,1400,800]
plt.pie(salary,labels=Company)
plt.show()
```

```python
plt.figure(figsize=(6,8))
plt.title("Pie_Chart")
Company=['Google','meta','msft','IBM']
salary=[1600,1200,1400,800]
plt.pie(salary,labels=Company,shadow=True,autopct='%.2f')
plt.show()
```

```python
plt.figure(figsize=(6,8))
plt.title("Pie_Chart")
Company=['Google','meta','msft','IBM']
salary=[1600,1200,1400,800]
plt.pie(salary,labels=Company,shadow=True,autopct='%.2f',explode=[0,0,0,0.2])
plt.show()
```
